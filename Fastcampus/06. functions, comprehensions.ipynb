```python
# Binet's Fibonacci formula

def fib_binet(n):
    sqrt_5 = math.sqrt(5)
    return (((1+sqrt_5)**n) -((1-sqrt_5)**n))/((2**n)*sqrt_5)
```

```python
def decorator_(function):
    def wrapper(*args, **kwargs):
        print('System : @decorator start')
        result = function(*args, **kwargs)
        print('System : @decorator finished')
        return 'success'
    return wrapper
```

```python
def decorator_(function):
    def wrapper(*args, **kwargs):
        print('System : @decorator start')
        result = function(*args, **kwargs)
        print('System : @decorator finished')
        return function(*args, **kwargs)
    return wrapper
```

```python
def time_checker(func):
    def wrapper(*args, **kargs):
        print('time checker start!')
        start_time = time.time()
        result = func(*args, **kargs)
        end_time = time.time()
        
        executed_time = end_time - start_time
        print('executed time for {name}: {time}s'.format(
            name = func.__name__,
            time = executed_time))
        return result
    return wrapper
```

```python
#동작 순서는 something->time_checker->decorator_순
@time_checker
@decorator_
def something(text):
    for i in range(2000):
        print('pass {}'.format(text))
    return 'something'
```

```python
### 도전과제 피보나치킨 ###

def f_serise(n):
    """
    n = 치킨을 먹을 사람의 숫자
    n보다 작거나 같은 값까지의 피보나치 수열을 생성하는 함수
    (피보나치 수열을 생성하지만 입력한 'n번째까지' 생성하는 것은 아님)
    """
    f = [1]
    if n <= 0:
        return print('1 이상의 숫자를 입력하세요')

    elif n == 1:
        return f
    
    elif n == 2:
        f.append(f[0])
        return f
    
    elif n >= 3:
        a,b = 1,1
        for i in range(1,n+1):
            a,b = b,a+b
            if n >= a:

                f.append(a)
        return f

def fibonachicken(n):
    f_list = f_serise(n) # f_series로 생긴 수열을 f_list라는 이름의 list로 받음 
    """
    Part 1
    n이 1,2,3일 경우는 default 1 출력(단, 0 이하인 경우 작동하지 않음)
    
    Part 2
    n이 피보나치 수열 x번째 값에 있을 경우, x-1마리 출력
    
    Part 3
    n이 피보니치 수열에 없을 경우,
    제켄도르프 정리를 통해 n을 분해하여 리스트로 저장 후,
    분해한 각 값의 최적 치킨 수의 합 출력
    """
    # Part 1
    if n <= 0:
        return print('1 이상의 숫자를 입력하세요')
    if n == 1 or n == 2:
        return 1 # default
    elif n == 3:
        return 2 # default
    
    # Part 2 : when n is in fibonacci series
    elif n in f_list:
        result = f_list[-2]

        return result
    
    # Part 2 : when n isn't in fibonacci series
    elif n not in f_list:
        f_list.reverse()
        print('flist',f_list)
        z_list = [] # z_list : a list of Zeckendorf's Theorem
        for i in f_list:
            x = f_list.index(i)
            y = f_list[x]
            print('y_list',y)

            z_list.append(y)
            if sum(z_list) > n:
                z_list.remove(y)

        cnum_list = [] # finding optimal chiken number of each Zeckendorf's number
        print('z:',z_list)
        for i in z_list: 
            x = f_list.index(i)
            print(x)
            y = f_list[x+1]
            print(y)
            cnum_list.append(y)
            print(cnum_list)
            result = sum(cnum_list)
        
        return result
```

```python
def tag_deco(function):
    def wrapper(*args, **kwargs):
        head = '<h1><b>'
        tail = '</h1></b>'
        return head + function(*args, **kwargs) + tail
    return wrapper
```

```python
def time_checker(function):
    def wrapper(*args, **kwargs):
        start = time.time()
        print('start: ', start)

        function(*args, **kwargs)
        end = time.time()
        print('end: ', end)        
        res = end - start
        print('result: ', res)
        return res
    return wrapper
```
